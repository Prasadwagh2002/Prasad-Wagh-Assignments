```python
#Generate a random array of 50 integers and display them using a line chart, scatter plot, histogram and box plot. Apply appropriate color, labels and styling options.
import numpy as np
import matplotlib.pyplot as plt

# Generate a random array of 50 integers between 1 and 100
np.random.seed(0)  # Setting seed for reproducibility
data = np.random.randint(1, 101, 50)

# Create a figure with 2x2 grid of subplots
plt.figure(figsize=(14, 10))

# Line Chart
plt.subplot(2, 2, 1)
plt.plot(data, color='blue', marker='o', linestyle='-', linewidth=2, markersize=5)
plt.title('Line Chart of Random Integers', fontsize=14)
plt.xlabel('Index', fontsize=12)
plt.ylabel('Value', fontsize=12)
plt.grid(True)

# Scatter Plot
plt.subplot(2, 2, 2)
plt.scatter(range(len(data)), data, color='green', s=50, edgecolor='black')
plt.title('Scatter Plot of Random Integers', fontsize=14)
plt.xlabel('Index', fontsize=12)
plt.ylabel('Value', fontsize=12)
plt.grid(True)

# Histogram
plt.subplot(2, 2, 3)
plt.hist(data, bins=10, color='orange', edgecolor='black')
plt.title('Histogram of Random Integers', fontsize=14)
plt.xlabel('Value', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True)

# Box Plot
plt.subplot(2, 2, 4)
plt.boxplot(data, patch_artist=True, boxprops=dict(facecolor='purple', color='black'), 
            medianprops=dict(color='yellow'), whiskerprops=dict(color='black'), 
            capprops=dict(color='black'))
plt.title('Box Plot of Random Integers', fontsize=14)
plt.ylabel('Value', fontsize=12)
plt.grid(True)

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the plots
plt.show()
```

```python
#1) Write a Python program to perform the following tasks : 
#a. Apply OneHot coding on Country column. 
#b. Apply Label encoding on purchased column 
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

# Sample DataFrame
data = {
    'Country': ['France', 'Spain', 'Germany', 'Spain', 'Germany', 'France', 'France', 'Spain', 'Germany', 'France'],
    'Purchased': ['No', 'Yes', 'No', 'No', 'Yes', 'Yes', 'No', 'Yes', 'No', 'Yes']
}

df = pd.DataFrame(data)

# Task 1a: Apply OneHot encoding on the 'Country' column
onehotencoder = OneHotEncoder(sparse=False)
country_encoded = onehotencoder.fit_transform(df[['Country']])

# Convert to DataFrame and add to original DataFrame
country_encoded_df = pd.DataFrame(country_encoded, columns=onehotencoder.get_feature_names_out(['Country']))
df = pd.concat([df, country_encoded_df], axis=1)

# Drop the original 'Country' column if you don't need it anymore
df.drop('Country', axis=1, inplace=True)

# Task 1b: Apply Label encoding on the 'Purchased' column
labelencoder = LabelEncoder()
df['Purchased'] = labelencoder.fit_transform(df['Purchased'])

# Print the updated DataFrame
print(df)
```

```python
#Write a python program to compute sum of Manhattan distance between all pairs of points.

import numpy as np

def manhattan_distance(point1, point2):
    return np.abs(point1[0] - point2[0]) + np.abs(point1[1] - point2[1])

def sum_of_manhattan_distances(points):
    total_distance = 0
    num_points = len(points)
    
    for i in range(num_points):
        for j in range(i + 1, num_points):
            total_distance += manhattan_distance(points[i], points[j])
    
    return total_distance

# Example usage:
points = [(1, 2), (4, 6), (7, 8), (2, 1)]  # Replace with your list of points
total_distance = sum_of_manhattan_distances(points)

print("Sum of Manhattan distances between all pairs of points:", total_distance)
```

```python
#Write a Python program to create data frame containing column name, salary, department add 10 rows with some missing and duplicate values to the data frame. Also drop all null and empty values. Print the modified data frame.


import pandas as pd
import numpy as np

# Create the initial DataFrame
data = {
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Edward', 'Alice', 'Frank', np.nan, 'Hannah', 'Ivy'],
    'salary': [50000, np.nan, 60000, 50000, 70000, 50000, 80000, 90000, np.nan, 100000],
    'department': ['HR', 'Engineering', np.nan, 'Engineering', 'HR', 'HR', 'Sales', 'Marketing', 'Sales', np.nan]
}

df = pd.DataFrame(data)

# Introduce duplicate rows
df = df.append(df.iloc[2], ignore_index=True)
df = df.append(df.iloc[7], ignore_index=True)

# Display the original DataFrame
print("Original DataFrame:")
print(df)

# Drop all rows with any null/empty values
df_cleaned = df.dropna().drop_duplicates()

# Display the modified DataFrame
print("\nModified DataFrame (After dropping nulls and duplicates):")
print(df_cleaned)
```

```python
#Write a Python NumPy program to compute the weighted average along the specified axis of a given flattened array.
import numpy as np

# Create a sample flattened array
array = np.array([1, 2, 3, 4, 5, 6])

# Create corresponding weights
weights = np.array([0.1, 0.2, 0.3, 0.2, 0.1, 0.1])

# Compute the weighted average along the specified axis
weighted_avg = np.average(array, weights=weights)

# Print the result
print("Weighted average of the array:", weighted_avg)
```

```python
#Write a python program to create two lists, one representing subject names and the other representing marks obtained in those subjects. Display the data in a pie chart and bar chart
import matplotlib.pyplot as plt

# Lists representing subject names and marks obtained
subjects = ['Math', 'Physics', 'Chemistry', 'Biology', 'English', 'History']
marks = [85, 90, 78, 92, 88, 76]

# Plotting the data in a pie chart
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.pie(marks, labels=subjects, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Marks Distribution in Subjects')

# Plotting the data in a bar chart
plt.subplot(1, 2, 2)
plt.bar(subjects, marks, color='skyblue')
plt.xlabel('Subjects')
plt.ylabel('Marks Obtained')
plt.title('Marks Obtained in Each Subject')

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the charts
plt.show()
```
